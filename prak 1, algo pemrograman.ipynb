```python
# Mendapatkan input dari pengguna
nama = input("Nama kamu siapa? ")
asal = input(f"Dari mana asal kamu, {nama}? ")
sekolah = input("Sebelumnya bersekolah di mana? ")

# Menampilkan hasil percakapan
print(f"\nHallo selamat siang {nama}")
print(f"Wah dari {asal}")
print(f"Pasti menyenangkan di {sekolah}")
```

```python
# Mendapatkan input dari pengguna
panjang = float(input("Masukkan Panjang Ruangan: "))
lebar = float(input("Masukkan Lebar Ruangan: "))
satuan = input("Masukkan Satuan (Meter/Inci): ")

# Menghitung luas ruangan
luas = panjang * lebar

# Menampilkan hasil perhitungan
print(f"\nLuas ruangan dengan panjang {panjang} dan lebar {lebar} adalah {luas} {satuan}")
```
